```python
import pandas as pd
import requests
import json
import os

apikey = os.environ.get('apikey')

def financialratios(symbol):
    fr = requests.get(f'https://financialmodelingprep.com/api/v3/financial-ratios/{symbol}?apikey={apikey}')
    fr = fr.json()
    valuation = fr['ratios'][0]['investmentValuationRatios']
    profitability = fr['ratios'][0]['profitabilityIndicatorRatios']
    operating = fr['ratios'][0]['operatingPerformanceRatios']
    liquidity = fr['ratios'][0]['liquidityMeasurementRatios']
    debt = fr['ratios'][0]['debtRatios']

    valuation = pd.DataFrame(list(valuation.items()),columns=['Ratio',symbol])
    profitability = pd.DataFrame(list(profitability.items()),columns=['Ratio',symbol])
    operating = pd.DataFrame(list(operating.items()),columns=['Ratio',symbol])
    liquidity = pd.DataFrame(list(liquidity.items()),columns=['Ratio',symbol])
    debt = pd.DataFrame(list(debt.items()),columns=['Ratio',symbol])

    frames = [valuation,profitability,operating,liquidity,debt]
    concatframes = pd.concat(frames)
    return concatframes

x = financialratios('GOOG')
listofsymbols = ['AAPL','MSFT','FB','TSLA']

for symbol in listofsymbols:
    y = financialratios(symbol)
    x = x.merge(y,on='Ratio')
x
```
