```python
# coefficient_of_variation = standard_deviation / mean

#https://www.thebalancesmb.com/how-to-calculate-business-risk-393472
import requests
import os
import numpy
apikey = os.environ.get('apikey')


stocks = ['AAPL','MSFT','GOOG']
for stock in stocks:
    income_statement = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{stock}?apikey={apikey}').json()
    #print(income_statement)
   
    number_of_years = 0
    revenue = []
    operating_income =[]
    net_income = []
    for item in income_statement:
        if number_of_years < 5:
            revenue.append(income_statement[number_of_years]['revenue'])
            operating_income.append(income_statement[number_of_years]['operatingIncome'])
            net_income.append(income_statement[number_of_years]['netIncome'])
            number_of_years += 1
    
    
    revenue_array = numpy.array(revenue)
    operating_income_array = numpy.array(operating_income)
    net_income_array = numpy.array(net_income)
    
    cv_revenue = revenue_array.std() / revenue_array.mean()
    print('Revenue coefficient variation for '+stock+' is '+str(round(cv_revenue,2)))
    
    cv_operating_income = operating_income_array.std() / operating_income_array.mean()
    print('Operating income coefficient variation for '+stock+' is '+str(round(cv_operating_income,2)))
    
    cv_net_income = net_income_array.std() / net_income_array.mean()
    print('Net income coefficient variation for '+stock+' is '+str(round(cv_net_income,2)))
    
```
