```python
import pandas as pd
import requests
import json
import os

apikey = os.environ.get('apikey')

def get_income_statement(symbol):
    r = requests.get(f'https://financialmodelingprep.com/api/v3/financials/income-statement/{symbol}?period=quarter&apikey={apikey}')
    r = r.json()
    incomeStatement = r['financials']
    incomeStatement = pd.DataFrame.from_dict(incomeStatement)
    incomeStatement = incomeStatement.T
    incomeStatement.columns = incomeStatement.iloc[0]
    incomeStatement.reset_index(inplace=True)
    incomeStatement = incomeStatement.iloc[:,0:2]
    incomeStatement.rename(columns={incomeStatement.columns[1]:symbol},inplace = True)
    cols = incomeStatement.columns.drop('index')
    incomeStatement[cols] = incomeStatement[cols].apply(pd.to_numeric,errors='coerce')
    incomeStatement = incomeStatement.iloc[1:,:]
    incomeStatement[symbol] = incomeStatement[symbol]/incomeStatement.iloc[0,1]  #*check details second trial
    incomeStatement[symbol] = pd.Series(['{0:.2f}%'.format(val*100) for val in incomeStatement[symbol]],index=incomeStatement.index)
    return incomeStatement

incomeStatements = get_income_statement('FB')
listOfSymbols = ['AMZN','AAPL','NFLX','GOOG']

for symbol in listOfSymbols:
    incomeStatement = get_income_statement(symbol)
    incomeStatements = incomeStatements.merge(incomeStatement,on='index')

incomeStatements
```
